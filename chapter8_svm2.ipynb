## فرمول ابرصفحه

با مجموعه نقاط، سوال بعدی یافتن فرمول ابرصفحه، همراه با دو حاشیه است. بدون پرداختن بیش از حد به ریاضیات پشت این موضوع، شکل 8.6 فرمول به دست آوردن ابرصفحه را نشان می‌دهد.
</div>

<div dir="rtl">
**شکل 8.6:** فرمول ابرصفحه و دو حاشیه همراه آن
</div>

<div style="display: flex; gap: 10px; justify-content: center; align-items: center;">

<div dir="rtl">
همانطور که در شکل 8.6 می‌بینید، فرمول ابرصفحه (g) به صورت زیر ارائه می‌شود:

$$\mathbf{g}(\mathbf{x}) = \tilde{W}_0\mathbf{x}_1 + \tilde{W}_1\mathbf{x}_2 + \mathbf{b}$$

که در آن \(\mathbf{x}_1\) و \(\mathbf{x}_2\) ورودی‌ها، \(\tilde{W}_0\) و \(\tilde{W}_1\) بردارهای وزن، و b بایاس است.

اگر مقدار g بزرگتر یا مساوی 1 باشد، نقطه مشخص شده در کلاس 1 است، و اگر مقدار g کوچکتر یا مساوی 1- باشد، نقطه مشخص شده در کلاس 2 است. همانطور که اشاره شد، هدف SVM یافتن عریض‌ترین حاشیه‌هایی است که کلاس‌ها را تقسیم می‌کند، و کل حاشیه (2d) به صورت زیر تعریف می‌شود:

$$2 / \|\mathbf{w}\|$$

که در آن \(\|\mathbf{w}\|\) بردارهای وزن نرمال شده \((\tilde{W}_0\) و \(\tilde{W}_1)\) است. با استفاده از مجموعه آموزش، هدف به حداقل رساندن مقدار \(\|\mathbf{w}\|\) است تا بتوانید حداکثر تفکیک‌پذیری بین کلاس‌ها را بدست آورید. پس از انجام این کار، قادر خواهید بود مقادیر \(\tilde{W}_0\)، \(\tilde{W}_1\) و b را بدست آورید.

یافتن حاشیه یک مسئله بهینه‌سازی محدود است که می‌تواند با استفاده از تکنیک ضرب‌کننده‌های لاگرانژ حل شود. بحث در مورد چگونگی یافتن حاشیه بر اساس مجموعه داده فراتر از محدوده این کتاب است، اما کافی است بگوییم که ما از کتابخانه Scikit-learn برای یافتن آنها استفاده خواهیم کرد.
</div>

## استفاده از Scikit-learn برای SVM

حال بیایید روی یک مثال کار کنیم تا ببینیم SVM چگونه کار می‌کند و چگونه می‌توان آن را با استفاده از Scikit-learn پیاده‌سازی کرد. برای این مثال، ما یک فایل به نام svm.csv داریم که حاوی داده‌های زیر است:

x1,x2,r
0,0,A
1,1,A
2,3,B
2,0,A
3,4,B

اولین کاری که انجام خواهیم داد رسم نقاط با استفاده از Seaborn است:
</div>

````python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns; sns.set(font_scale=1.2)
import matplotlib.pyplot as plt

data = pd.read_csv('svm.csv')
sns.lmplot(data=data, 
           x='x1', 
           y='x2',
           hue='r',
           palette='Set1',
           fit_reg=False,
           scatter_kws={"s": 50})
````

<div dir="rtl">
**شکل 8.7:** نقاط رسم شده با استفاده از Seaborn
</div>

<div dir="rtl">
با استفاده از نقاط داده‌ای که قبلاً بارگذاری کردیم، حال بیایید از کلاس svc ماژول svm کتابخانه Scikit-learn استفاده کنیم تا به ما در استخراج مقدار متغیرهای مختلف که در غیر این صورت باید محاسبه کنیم، کمک کند. قطعه کد زیر از کرنل خطی برای حل مسئله استفاده می‌کند. کرنل خطی فرض می‌کند که مجموعه داده را می‌توان به صورت خطی جدا کرد.
</div>

````python
from sklearn import svm
#---Converting the Columns as Matrices---
points = data[['x1','x2']].values
result = data['r']

clf = svm.SVC(kernel = 'linear')
clf.fit(points, result)

print('Vector of weights (w) = ',clf.coef_[0])
print('b = ',clf.intercept_[0])
print('Indices of support vectors = ', clf.support_)
print('Support vectors = ', clf.support_vectors_)
print('Number of support vectors for each class = ', clf.n_support_)
print('Coefficients of the support vector in the decision function = ',
       np.abs(clf.dual_coef_))
````

<div dir="rtl">
کلاس SVC مخفف Support Vector Classification است. ماژول svm شامل مجموعه‌ای از کلاس‌ها است که SVM را برای اهداف مختلف پیاده‌سازی می‌کنند:

- svm.LinearSVC: طبقه‌بندی بردار پشتیبان خطی
- svm.LinearSVR: رگرسیون بردار پشتیبان خطی
- svm.NuSVC: طبقه‌بندی بردار پشتیبان Nu
- svm.NuSVR: رگرسیون بردار پشتیبان Nu
- svm.OneClassSVM: تشخیص ناهنجاری بدون نظارت
- svm.SVC: طبقه‌بندی بردار پشتیبان C
- svm.SVR: رگرسیون بردار پشتیبان اپسیلون

**نکته:** برای این فصل، تمرکز ما بر استفاده از SVM برای طبقه‌بندی است، حتی اگر SVM می‌تواند برای رگرسیون نیز استفاده شود.

قطعه کد قبلی خروجی زیر را تولید می‌کند:

## رسم ابرصفحه و حاشیه‌ها

با تمام مقادیر متغیرها، اکنون زمان آن رسیده است که ابرصفحه و دو حاشیه همراه آن را رسم کنیم. آیا فرمول ابرصفحه را به یاد دارید؟ به صورت زیر است:

$$\mathbf{g}(\mathbf{x}) = \tilde{W}_0\mathbf{x}_1 + \tilde{W}_1\mathbf{x}_2 + \mathbf{b}$$

برای رسم ابرصفحه، \(\tilde{W}_0\mathbf{x}_1 + \tilde{W}_1\mathbf{x}_2 + \mathbf{b}\) را برابر 0 قرار دهید:

$$\tilde{W}_0\mathbf{x}_1 + \tilde{W}_1\mathbf{x}_2 + \mathbf{b} = 0$$

برای رسم ابرصفحه (که در این حالت یک خط مستقیم است)، به دو نقطه نیاز داریم: یکی روی محور x و یکی روی محور y.

با استفاده از فرمول قبلی، وقتی \(\mathbf{x}_1 = 0\) است، می‌توانیم \(\mathbf{x}_2\) را به صورت زیر حل کنیم:

$$\tilde{W}_1\mathbf{x}_2 + \mathbf{b} = 0 \implies \mathbf{x}_2 = -\frac{\mathbf{b}}{\tilde{W}_1}$$

وقتی \(\mathbf{x}_2 = 0\) است، می‌توانیم \(\mathbf{x}_1\) را به صورت زیر حل کنیم:

$$\tilde{W}_0\mathbf{x}_1 + \mathbf{b} = 0 \implies \mathbf{x}_1 = -\frac{\mathbf{b}}{\tilde{W}_0}$$

نقطه \((0, -\mathbf{b} / \tilde{W}_1)\) نقطه برخورد خط با محور y است. شکل 8.9 دو نقطه را روی دو محور نشان می‌دهد.
</div>

<div dir="rtl">
**شکل 8.9:** دو نقطه برخورد برای ابرصفحه
</div>

<div style="display: flex; gap: 10px; justify-content: center; align-items: center;">
  <img src="8-9.png" width="700" height="450">
</div>

<div dir="rtl">
هنگامی که نقاط روی هر محور پیدا شد، اکنون می‌توانید شیب را به صورت زیر محاسبه کنید:
</div>

````python
#---w is the vector of weights---
w = clf.coef_[0]

#---find the slope of the hyperplane---
slope = -w[0] / w[1]

b = clf.intercept_[0]

#---find the coordinates for the hyperplane---
xx = np.linspace(0, 4)
yy = slope * xx - (b / w[1])

#---plot the margins---
s = clf.support_vectors_[0]    #---first support vector---
yy_down = slope * xx + (s[1] - slope * s[0])

s = clf.support_vectors_[-1]   #---last support vector---
yy_up   = slope * xx + (s[1] - slope * s[0])

#---plot the points---
sns.lmplot(data=data, x='x1', y='x2', hue='r', palette='Set1', fit_reg=False, scatter_kws={"s": 50})

#---plot the hyperplane---
plt.plot(xx, yy, linewidth=2, color='green');

#---plot the 2 margins---
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
````

## افزودن بعد سوم

برای انجام این کار، می‌توانیم یک بعد سوم، مثلاً محور z را اضافه کنیم و z را به صورت زیر تعریف کنیم:

$$z = x^{2} + y^{2}$$

هنگامی که نقاط را با استفاده از نمودار سه بعدی رسم می‌کنیم، نقاط اکنون به صورت خطی قابل تفکیک هستند. تجسم این موضوع بدون رسم نقاط دشوار است. قطعه کد زیر دقیقاً این کار را انجام می‌دهد:
</div>

````python
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles

#---X is features and c is the class labels---
X, c = make_circles(n_samples=500, noise=0.09)

rgb = np.array(['r', 'g'])
plt.scatter(X[:, 0], X[:, 1], color=rgb[c])
plt.show()

fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')
z = X[:,0]**2 + X[:,1]**2
ax.scatter(X[:, 0], X[:, 1], z, color=rgb[c])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

## رسم ابرصفحه سه بعدی

با رسم نقاط در یک نمودار سه بعدی، حال بیایید مدل را با استفاده از بعد سوم آموزش دهیم:
</div>

````python
#---combine X (x-axis,y-axis) and z into single ndarray---
features = np.concatenate((X,z.reshape(-1,1)), axis=1)

#---use SVM for training---
from sklearn import svm

clf = svm.SVC(kernel = 'linear')
clf.fit(features, c)
````

<div dir="rtl">
ابتدا سه محور را با استفاده از تابع np.concatenate() در یک آرایه ndarray ترکیب کردیم. سپس مدل را طبق معمول آموزش دادیم. برای مجموعه نقاط قابل تفکیک خطی در دو بعد، فرمول ابرصفحه به صورت زیر است:

$$g(x) = \tilde{W}_{0x1} + \tilde{W}_{1x2} + b$$

برای مجموعه نقاطی که اکنون در سه بعد هستند، فرمول اکنون به صورت زیر می‌شود:

$$g(x) = \tilde{W}_{0x1} + \tilde{W}_{1x2} + \tilde{W}_{2x3} + b$$

به طور خاص، \(\tilde{W}_2\) اکنون توسط clf.coef_[0][2] نشان داده می‌شود، همانطور که در شکل 8.14 نشان داده شده است.
</div>

<div dir="rtl">
مرحله بعدی رسم ابرصفحه در سه بعد است. برای انجام این کار، باید مقدار x3 را پیدا کنید، که می‌تواند به صورت نشان داده شده در شکل 8.15 استخراج شود.
</div>

<div dir="rtl">
**شکل 8.15:** فرمول یافتن ابرصفحه در سه بعد

<div dir="rtl">
این را می‌توان در کد به صورت زیر بیان کرد:

```python
x3 = lambda x,y: (-clf.intercept_[0] - clf.coef_[0][0] * x - clf.coef_[0][1] * y) / clf.coef_[0][2]
```

برای رسم ابرصفحه در سه بعد، از تابع plot_surface() استفاده کنید:
</div>

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles

#---X is features and c is the class labels---
X, c = make_circles(n_samples=500, noise=0.09)
z = X[:,0]**2 + X[:,1]**2

rgb = np.array(['r', 'g'])

fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], z, color=rgb[c])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.show()

#---combine X (x-axis,y-axis) and z into single ndarray---
features = np.concatenate((X,z.reshape(-1,1)), axis=1)

#---use SVM for training---
from sklearn import svm

clf = svm.SVC(kernel = 'linear')
clf.fit(features, c)
x3 = lambda x,y: (-clf.intercept_[0] - clf.coef_[0][0] * x - clf.coef_[0][1] * y) / \
                   clf.coef_[0][2]

tmp = np.linspace(-1.5,1.5,100)
x,y = np.meshgrid(tmp,tmp)

ax.plot_surface(x, y, x3(x,y))
plt.show()
````
