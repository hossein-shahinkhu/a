## استفاده از مجموعه داده Scikit-learn

Scikit-learn با چند مجموعه داده نمونه استاندارد ارائه می‌شود که یادگیری یادگیری ماشین را آسان می‌کند. برای بارگذاری مجموعه داده‌های نمونه، ماژول `datasets` را وارد کرده و مجموعه داده مورد نظر را بارگذاری کنید. به عنوان مثال، قطعه کد زیر مجموعه داده Iris را بارگذاری می‌کند:

**نکته:** مجموعه داده گل Iris یا مجموعه داده فیشر، یک مجموعه داده چندمتغیره است که توسط آماردان و زیست‌شناس بریتانیایی، رونالد فیشر معرفی شده است. این مجموعه داده شامل ۵۰ نمونه از هر یک از سه گونه Iris (Iris setosa، Iris virginica و Iris versicolor) است. چهار ویژگی از هر نمونه اندازه‌گیری شده است: طول و عرض کاسبرگ‌ها و گلبرگ‌ها بر حسب سانتی‌متر. بر اساس ترکیب این چهار ویژگی، فیشر یک مدل تفکیک خطی برای تشخیص گونه‌ها از یکدیگر توسعه داد.

## مجموعه داده توزیع شده خطی

تابع `make_regression()` داده‌هایی را تولید می‌کند که به صورت خطی توزیع شده‌اند. می‌توانید تعداد ویژگی‌های مورد نظر خود و همچنین انحراف معیار نویز گاوسی اعمال شده به خروجی را مشخص کنید.

</div>

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets._samples_generator import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=5.4)
plt.scatter(X,y)
```

<div dir="rtl">

## مجموعه داده خوشه‌ای

تابع `make_blobs()` تعداد n خوشه از داده‌های تصادفی تولید می‌کند. این برای انجام خوشه‌بندی در یادگیری بدون نظارت بسیار مفید است (فصل 9، "یادگیری نظارت شده - طبقه‌بندی با استفاده از K نزدیک‌ترین همسایه (KNN)").

</div>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

X, y = make_blobs(500, centers=4)  # Generate isotropic Gaussian
                                   # blobs for clustering

rgb = np.array(['r', 'g', 'b', 'y'])

# plot the blobs using a scatter plot and use color coding
plt.scatter(X[:, 0], X[:, 1], color=rgb[y])
```

# شروع کار با Scikit-learn

ساده‌ترین راه برای شروع کار با یادگیری ماشین با Scikit-learn، شروع با رگرسیون خطی است. رگرسیون خطی یک رویکرد خطی برای مدل‌سازی رابطه بین یک متغیر وابسته اسکالر y و یک یا چند متغیر توضیحی (یا متغیرهای مستقل) است. به عنوان مثال، تصور کنید که مجموعه داده‌ای شامل قد (بر حسب متر) یک گروه از افراد و وزن‌های متناظر آنها (بر حسب کیلوگرم) دارید.

</div>

```python
%matplotlib inline
import matplotlib.pyplot as plt

# represents the heights of a group of people in metres
heights = [[1.6], [1.65], [1.7], [1.73], [1.8]]

# represents the weights of a group of people in kgs
weights = [[60], [65], [72.3], [75], [80]]

plt.title('Weights plotted against heights')
plt.xlabel('Heights in metres')
plt.ylabel('Weights in kilograms')

plt.plot(heights, weights, 'k.')

# axis range for x and y
plt.axis([1.5, 1.85, 50, 90])
plt.grid(True)
```

<div dir="rtl">

از نمودار، می‌توانید ببینید که بین وزن و قد برای این گروه از افراد همبستگی مثبت وجود دارد. می‌توانید یک خط مستقیم از میان نقاط رسم کنید و از آن برای پیش‌بینی وزن فرد دیگر بر اساس قد او استفاده کنید.

</div>

<div dir="rtl">

## استفاده از کلاس LinearRegression برای برازش مدل

پس چگونه خط مستقیمی را رسم می‌کنیم که از میان همه نقاط عبور کند؟ معلوم می‌شود که کتابخانه Scikit-learn کلاس `LinearRegression` را دارد که به شما کمک می‌کند دقیقاً همین کار را انجام دهید. تنها کاری که باید انجام دهید این است که یک نمونه از این کلاس ایجاد کنید و از لیست‌های heights و weights برای ایجاد یک مدل رگرسیون خطی با استفاده از تابع `fit()` استفاده کنید، مانند این:

**نکته:** توجه داشته باشید که heights و weights هر دو به صورت لیست‌های دو بعدی نشان داده شده‌اند. تابع `fit()` به هر دو آرگومان X و y نیاز دارد که دو بعدی باشند (از نوع list یا ndarray).

</div>

```python
from sklearn.linear_model import LinearRegression

# Create and fit the model
model = LinearRegression()
model.fit(X=heights, y=weights)
```

<div dir="rtl">

## انجام پیش‌بینی

هنگامی که مدل را برازش (آموزش) دادید، می‌توانید با استفاده از تابع `predict()` پیش‌بینی را شروع کنید، مانند این:

در مثال زیر، می‌خواهید وزن فردی با قد 1.75 متر را پیش‌بینی کنید. بر اساس مدل، وزن 76.04 کیلوگرم پیش‌بینی می‌شود.

**نکته:** در Scikit-learn، معمولاً از تابع `fit()` برای آموزش یک مدل استفاده می‌کنید. پس از آموزش مدل، از تابع `predict()` برای انجام پیش‌بینی استفاده می‌کنید.

</div>

```python
# make prediction
weight = model.predict([[1.75]])[0][0]
print(round(weight,2))         # 76.04
```

```python
print(model.predict([[1.75],[1.35],[1.44]])[0][0])
```

<div dir="rtl">

## رسم خط رگرسیون خطی

تجسم خط رگرسیون خطی که توسط کلاس `LinearRegression` ایجاد شده است مفید خواهد بود. بیایید این کار را با اول رسم نقاط داده اصلی و سپس ارسال لیست heights به مدل برای پیش‌بینی weights انجام دهیم. سپس سری weights پیش‌بینی شده را رسم می‌کنیم تا خط را بدست آوریم.

</div>

```python
import matplotlib.pyplot as plt

heights = [[1.6], [1.65], [1.7], [1.73], [1.8]]
weights = [[60], [65], [72.3], [75], [80]]

plt.title('Weights plotted against heights')
plt.xlabel('Heights in metres')
plt.ylabel('Weights in kilograms')

plt.plot(heights, weights, 'k.')

plt.axis([1.5, 1.85, 50, 90])
plt.grid(True)

# plot the regression line
plt.plot(heights, model.predict(heights), color='r')
```

<div dir="rtl">

## دریافت شیب و عرض از مبدأ خط رگرسیون خطی

از نمودار، مشخص نیست که خط رگرسیون خطی در چه مقداری محور y را قطع می‌کند. محور x را طوری تنظیم کرده‌ایم که رسم از 1.5 شروع شود. راه بهتر برای تجسم این است که محور x را طوری تنظیم کنیم که از 0 شروع شود و محدوده محور y را بزرگتر کنیم. سپس خط را با وارد کردن دو مقدار حدی height: 0 و 1.8 رسم می‌کنید.

</div>

```python
plt.title('Weights plotted against heights')
plt.xlabel('Heights in metres')
plt.ylabel('Weights in kilograms')

plt.plot(heights, weights, 'k.')

plt.axis([0, 1.85, -200, 200])
plt.grid(True)

# plot the regression line
extreme_heights = [[0], [1.8]]
plt.plot(extreme_heights, model.predict(extreme_heights), color='b')
```

## بررسی عملکرد مدل با محاسبه مجموع مربعات باقیمانده

برای اینکه بدانید خط رگرسیون خطی شما به خوبی بر روی همه نقاط داده برازش شده است یا خیر، از روش مجموع مربعات باقیمانده (RSS) استفاده می‌کنیم.

## ارزیابی مدل با استفاده از مجموعه داده تست

اکنون که مدل ما با داده‌های آموزش ما آموزش دیده است، می‌توانیم آن را آزمایش کنیم. با فرض اینکه مجموعه داده تست زیر را داریم:

</div>

```python
# test data
heights_test = [[1.58], [1.62], [1.69], [1.76], [1.82]]
weights_test = [[58], [63], [72], [73], [85]]
```

<div dir="rtl">

می‌توانیم میزان نزدیکی داده‌های تست به خط رگرسیون را با استفاده از روش R-Squared اندازه‌گیری کنیم. روش R-Squared به عنوان ضریب تعیین، یا ضریب تعیین چندگانه برای رگرسیون‌های چندگانه نیز شناخته می‌شود.

با استفاده از فرمول نشان داده شده برای R-Squared، به نکات زیر توجه کنید:
- R^2 همان R-squared است
- TSS مجموع مربعات کل است
- RSS مجموع مربعات باقیمانده است

**نکته:** تابع `ravel()` لیست دو بعدی را به یک آرایه مسطح (یک بعدی) پیوسته تبدیل می‌کند.

</div>

```python
# Total Sum of Squares (TSS)
weights_test_mean = np.mean(np.ravel(weights_test))
TSS = np.sum((np.ravel(weights_test) -
              weights_test_mean) ** 2)
print("TSS: %.4f" % TSS)

# Residual Sum of Squares (RSS)
RSS = np.sum((np.ravel(weights_test) -
              np.ravel(model.predict(heights_test)))
                 ** 2)
print("RSS: %.4f" % RSS)

# R_squared
R_squared = 1 - (RSS / TSS)
print("R-squared: %.4f" % R_squared)
```

<div dir="rtl">

خوشبختانه، نیازی نیست R-Squared را به صورت دستی محاسبه کنید - Scikit-learn تابع `score()` را دارد که به طور خودکار R-Squared را برای شما محاسبه می‌کند:

</div>

```python
# using scikit-learn to calculate r-squared
print('R-squared: %.4f' % model.score(heights_test,
                                      weights_test))

# R-squared: 0.9429
```

<div dir="rtl">

مقدار R-Squared برابر با 0.9429 (94.29٪) نشان‌دهنده برازش بسیار خوب برای داده‌های تست شما است.

</div>

## حذف نقاط پرت

در آمار، یک نقطه پرت، نقطه‌ای است که از سایر نقاط مشاهده شده دور است. به عنوان مثال، با توجه به مجموعه مقادیر 

- 234، 267، 1، 200، 245، 300، 199، 250، 8999، 245

کاملاً واضح است که 1 و 8999 نقاط پرت هستند. آنها به وضوح از بقیه مقادیر متمایز هستند، و "خارج از" بیشتر مقادیر دیگر در مجموعه داده قرار می‌گیرند؛ از این رو کلمه outlier (نقطه پرت) به کار می‌رود. نقاط پرت عمدتاً به دلیل خطاهای ثبت یا خطای آزمایشی رخ می‌دهند، و در یادگیری ماشین حذف آنها قبل از آموزش مدل مهم است، زیرا اگر حذف نشوند، ممکن است مدل شما را مخدوش کنند.

تعدادی تکنیک برای حذف نقاط پرت وجود دارد، و در این فصل دو مورد از آنها را بحث می‌کنیم:
- نرده‌های توکی
- نمره Z

</div>

<div dir="rtl">

### نرده‌های توکی

نرده‌های توکی بر اساس دامنه بین چارکی (IQR) است. IQR تفاوت بین چارک اول و سوم یک مجموعه مقادیر است. چارک اول، که با Q1 نشان داده می‌شود، مقداری در مجموعه داده است که 25٪ از مقادیر زیر آن قرار دارد. چارک سوم، که با Q3 نشان داده می‌شود، مقداری در مجموعه داده است که 25٪ از مقادیر بالای آن قرار دارد. از این رو، طبق تعریف، IQR = Q3 - Q1.

</div>
<div style="display: flex; gap: 10px; justify-content: center; align-items: center;">
  <img src="IQR.png" width="700" height="450">
</div>


<div dir="rtl">

در نرده‌های توکی، نقاط پرت مقادیری هستند که:
- کمتر از Q1 - (1.5 × IQR) یا
- بیشتر از Q3 + (1.5 × IQR)

</div>


<div dir="rtl">

قطعه کد زیر پیاده‌سازی نرده‌های توکی را با استفاده از پایتون نشان می‌دهد.

</div>


```python
import numpy as np

def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.3)
    upper_bound = q3 + (iqr * 1.3)
    return np.where((data > upper_bound) | (data < lower_bound))
```

<div dir="rtl">

**نکته:** تابع `np.where()` مکان آیتم‌هایی را که شرایط را برآورده می‌کنند، برمی‌گرداند.

تابع `outliers_iqr()` یک تاپل برمی‌گرداند که اولین عنصر آن آرایه‌ای از ایندکس‌های آن سطرهایی است که مقادیر پرت دارند.

</div>

### نمره Z

روش دوم برای تعیین نقاط پرت، استفاده از روش نمره Z است. نمره Z نشان می‌دهد که یک نقطه داده چند انحراف معیار از میانگین فاصله دارد. نمره Z فرمول زیر را دارد:

Z = (xi - μ) / σ

که در آن xi نقطه داده، μ میانگین مجموعه داده، و σ انحراف معیار است.

نحوه تفسیر نمره Z به این صورت است:
- نمره Z منفی نشان می‌دهد که نقطه داده کمتر از میانگین است، و نمره Z مثبت نشان می‌دهد که نقطه داده مورد نظر بزرگتر از میانگین است
- نمره Z 0 به شما می‌گوید که نقطه داده دقیقاً در وسط (میانگین) قرار دارد، و نمره Z 1 به شما می‌گوید که نقطه داده شما 1 انحراف معیار بالاتر از میانگین است، و غیره
- هر نمره Z بزرگتر از 3 یا کمتر از 3- به عنوان یک نقطه پرت در نظر گرفته می‌شود

قطعه کد زیر پیاده‌سازی نمره Z را با استفاده از پایتون نشان می‌دهد.

با استفاده از همان مجموعه داده Galton که قبلاً استفاده کردید، اکنون می‌توانید نقاط پرت را برای ستون height با استفاده از تابع `outliers_z_score()` پیدا کنید.

با استفاده از روش نمره Z، می‌توانید ببینید که ستون height دارای سه نقطه پرت است.

</div>

```python
def outliers_z_score(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(y - mean) / std for y in data]
    return np.where(np.abs(z_scores) > threshold)
```

```python
print("Outliers using outliers_z_score()")
print("=================================")
for i in outliers_z_score(df.childHeight)[0]:
    print(df[i:i+1])
print()
```
